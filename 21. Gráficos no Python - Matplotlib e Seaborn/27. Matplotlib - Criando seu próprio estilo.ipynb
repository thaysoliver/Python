## Criando Seu Próprio Estilo

Embora o Matplotlib venha com uma variedade de estilos pré-prontos, você também pode criar e usar seu próprio estilo personalizado. Isso pode ser útil se você quiser que seus gráficos tenham uma aparência específica que não é fornecida por nenhum dos estilos pré-prontos.

Para criar seu próprio estilo, você precisa criar um arquivo de configuração que define as configurações de estilo que você deseja usar. Este arquivo deve ser um arquivo `.mplstyle`, que é basicamente um arquivo de texto que contém uma série de configurações no formato `chave: valor`.

Por exemplo, aqui está um exemplo de um arquivo de estilo que define algumas configurações básicas:

```
axes.titlesize : 24
axes.labelsize : 20
xtick.labelsize : 16
ytick.labelsize : 16
```

Este arquivo define o tamanho do título e do rótulo dos eixos, a largura da linha, o tamanho do marcador e o tamanho do rótulo do tick.



````python
import matplotlib.pyplot as plt

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]

barras = plt.barh(produtos, quantidades)

plt.bar_label(barras, labels=quantidades, padding=3, fontsize=10, fontweight='bold')

plt.xticks([])
plt.title('Quantidade de Produtos Vendidos')
plt.grid(False)

plt.show()
````

````python
import matplotlib.pyplot as plt

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]

plt.style.use('meu_estilo_1.mplstyle')

barras = plt.barh(produtos, quantidades)

plt.bar_label(barras, labels=quantidades, padding=3, fontsize=10, fontweight='bold')

plt.xticks([])
plt.title('Quantidade de Produtos Vendidos')
plt.grid(False)

plt.show()
````
