# API Orientada a Objetos

## Introdução à API Orientada a Objetos e por que ela é útil

A API Orientada a Objetos do Matplotlib é uma abordagem alternativa à API `pyplot` que oferece maior controle sobre a criação e personalização de gráficos. Em vez de depender de um estado global, a API Orientada a Objetos trabalha diretamente com figuras e eixos. Isso é especialmente útil quando você quer criar gráficos mais complexos, como figuras com vários subplots.

## Limitações do plt e como a API Orientada a Objetos supera essas limitações

A API `pyplot` é ótima para gráficos simples e rápidos, mas pode ser menos flexível quando se trata de criar gráficos mais complexos. Por exemplo, criar subplots ou figuras com vários eixos pode ser complicado com `pyplot`.

A API Orientada a Objetos, por outro lado, dá mais controle sobre seus gráficos. Você pode trabalhar diretamente com instâncias de `Figure` e `Axes`, permitindo que você manipule cada aspecto do seu gráfico. Isso é especialmente útil para gráficos mais complexos e para a criação de visualizações mais personalizadas.
